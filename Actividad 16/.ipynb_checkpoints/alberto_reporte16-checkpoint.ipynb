```python
import random
import datetime

def experimento(N, R):
    resultados = []
    for n in range(2, N):
        for r in range(R):
            objetos = []
            for i in range(n):
                objeto = {
                    "id" : i,
                    "peso" : random.random(),
                    "valor" : random.random()
                }
                objetos.append(objeto)

            peso_total = 0
            valor_total = 0
            for i in range(n):
                peso_total += objetos[i]['peso']
                valor_total += objetos[i]['valor']

            peso_sup = N / 4

            inicio = datetime.datetime.now()
            peso_mejor = 0
            valor_mejor = 0
            mejor_subconjunto = ""
            
            for i in range(2 ** n):
                # https://stackoverflow.com/a/3252544
                subconjunto = ('{0:0' + str(n) + 'b}').format(i)
                subconjunto_arr = []
                peso_local = 0
                valor_local = 0
                for j in range(len(subconjunto)):
                    if subconjunto[j] == '1':
                        if peso_local + objetos[j]['peso'] <= peso_sup:
                            peso_local += objetos[j]['peso']
                            valor_local += objetos[j]['valor']
                            subconjunto_arr.append(objetos[j]['id'])
                        else:
                            break
                if valor_local >= valor_mejor:
                    valor_mejor = valor_local
                    peso_mejor = peso_local
                    mejor_subconjunto = subconjunto_arr
            tiempo_mejores = datetime.datetime.now() - inicio


            inicio = datetime.datetime.now()
            # Ranking valor / peso
            for i in range(n):
                objetos[i]['rank'] = objetos[i]['valor'] / objetos[i]['peso']

            # https://stackoverflow.com/a/73050
            ranked = None
            ranked = sorted(objetos, key = lambda k: k['rank'])  

            ranked = ranked[::-1]

            peso_aprox = 0
            valor_aprox = 0
            mejor_aprox = []

            for i in range(n):
                if peso_aprox + ranked[i]['peso'] <= peso_sup:
                    peso_aprox += ranked[i]['peso']
                    valor_aprox += ranked[i]['valor']
                    mejor_aprox.append(ranked[i]['id'])
                else:
                    continue

            tiempo_aprox = datetime.datetime.now() - inicio

            resultado = {}
            resultado['n'] = n
            resultado['r'] = r

            resultado['peso_sup'] = peso_sup
            resultado['peso_total'] = peso_total
            resultado['valor_total'] = valor_total

            resultado['mejor_subconjunto'] = mejor_subconjunto
            resultado['valor_mejor'] = valor_mejor
            resultado['peso_mejor'] = peso_mejor
            resultado['tiempo_mejores'] = tiempo_mejores

            resultado['mejor_aprox'] = mejor_aprox
            resultado['valor_aprox'] = valor_aprox
            resultado['peso_aprox'] = peso_aprox
            resultado['tiempo_aprox'] = tiempo_aprox

            resultados.append(resultado)
        
    return resultados
```

```python
experimentos = experimento(25, 10)
```

```python
for i in range(len(experimentos)):
    experimentos[i]['valor_dif'] = experimentos[i]['valor_aprox'] / experimentos[i]['valor_mejor']
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame(experimentos)
display(df)
```

```python
fig, ax = plt.subplots(figsize=(10,4))
df.boxplot(column=['tiempo_mejores'], by='n', ax = ax)
ax.grid(b=None)
plt.ylabel('Tiempo (s)', fontsize=18)
plt.xlabel('Objetos', fontsize=18)
plt.title('')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.suptitle('')
plt.show()
plt.close()

fig, ax = plt.subplots(figsize=(10,4))
df.boxplot(column=['valor_dif'], by='n', ax = ax)
ax.grid(b=None)
plt.title('')
plt.suptitle('')
plt.xlabel('Objetos', fontsize=18)
plt.ylabel('Similitud con Ã³ptimo', fontsize=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
plt.close()
```
